```python
## Q.1) How many attacks happened in India?

A1 = terror.query("country_txt == 'India'")['attacktype1'].count()
A2= terror.query("country_txt == 'India'")['attacktype2'].count()
A3= terror.query("country_txt == 'India'")['attacktype3'].count()
Total_Attack = A1 + A2 + A3   
Total_Attack                                         
```

```python
## Q.2) How many attacks happened in India and upto 3 people were killed?

D1 = terror[(terror.nkill<=3) & (terror.country_txt == 'India')]['attacktype1'].count()
D2 = terror[(terror.nkill<=3) & (terror.country_txt == 'India')]['attacktype2'].count()
D3 = terror[(terror.nkill<=3) & (terror.country_txt == 'India')]['attacktype3'].count()
Totalwithkillupto3 = D1 + D2 + D3
Totalwithkillupto3
```

```python
##Q.7) Label all the incidents where the number killed was more than 5 as severe.
#Hint:
#Use Aggregations and manipulations using apply and map
#map: map a function to each element of a series object
#Suppose we want to label all the incidents where the number killed was more
#than 5 as severe. This would involve applying a function on each element of the
#series, map helps in doing that.
#You can use lambda functions as well (if else follows a special form when used
#with lambdas )
#Use the nkill column.

terrorsevere = (terror['nkill'] > 5).map({True:'severe', False:'not severe'})
print(terrorsevere)
```

```python
##Q.8) write a function to label an incident that was both successful and suicidal
    
def label_incident(row):
    if row['success']==1 and row['suicide']==1:
        return 1
    else:
        return 0
    
```

```python
##Q.9) Create a new category representing if the incident occured in Afghanistan,
##Pakistan or India as one level of the category and all the other countries as
##another level. Label all other countries as ROW and new column which contains
##the new category as ‘Local’ in the data.
##Hint: Create a new category representing if the incident occured in Afghanistan,
##Pakistan or India as one level of the category and all the other countries as
##another level.
##Use the columns – country_txt and create a new column called “Local” in the
##dataset.


terror['Local'] = terror.apply(lambda row: 'Local' if row['country_txt']=='India' or row['country_txt']=='Pakistan' or row['country_txt']=='Afghanistan' else 'ROW',axis=1)
terror
```

```python
##Q.10) How many incidents happened in Af-Pak-India vs ROW?
##Hint: Use the newly created column “Local” and do a value_count or use the size
##function.


terror['Local'].value_counts()
```

```python
##Q.11) List the number of suicides attacks and average kills by Af-Pak-India vs
##ROW. Rename columns in the output as Average_Kills for average kills and
##Number_Incidents for count of suicide attacks.
##Hint: Use the newly created Local column and Suicide column for group by
##summary and use the nkill column to take average and count.


terror.groupby('Local', as_index = False).agg({'suicide': np.sum, 'nkill': np.mean}).rename(columns={"suicide":"Number_Incidents","nkill":"Average_Kills"})
```
